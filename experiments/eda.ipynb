```python
from langdetect import detect, DetectorFactory

DetectorFactory.seed = 0

```

```python
col_mapping = {
    'id': 'comment_id',
    'snippet.topLevelComment.snippet.authorDisplayName': 'author',
    'snippet.topLevelComment.snippet.textOriginal': 'text',
    'snippet.topLevelComment.snippet.likeCount': 'likes'
}
```

```python
df_clean = df[list(col_mapping.keys())].rename(columns=col_mapping)
```

```python
df_clean = df_clean.dropna(subset=['text'])
```

```python
def is_english(text):
    try:
        # returns 'en' for English
        return detect(text) == 'en'
    except Exception:
        # Safe fallback for entries containing only emojis/punctuation
        return False
# 4. Filter for English comments only
print(f"Total comments before filtering: {len(df_clean)}")
df_clean = df_clean[df_clean['text'].apply(is_english)]
print(f"Total English comments after filtering: {len(df_clean)}")
```
