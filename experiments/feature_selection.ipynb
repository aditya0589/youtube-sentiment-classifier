```python
df_final.rename(columns={
    'snippet.topLevelComment.snippet.textOriginal': 'text',
    'snippet.topLevelComment.snippet.likeCount': 'likes',
    'snippet.topLevelComment.id': 'comment_id'
}, inplace=True)
df_final.head()
```

```python
def get_vader_sentiment(text):
    if pd.isna(text) or str(text).strip() == "":
        return 'neutral'
    
    # Get scores dict: {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
    scores = sia.polarity_scores(str(text))
    compound = scores['compound']
    
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
df_final['sentiment'] = df_final['text'].apply(get_vader_sentiment)
```
